```python
import numpy as np 
import h5py
import matplotlib.pyplot as plt 
import matplotlib
%matplotlib inline
```

```python
filename = "../data/incommensurate_susc/n0.82/suscBZ.txt"
suBZ = np.loadtxt(filename)
ax = plt.subplot()
cf = ax.pcolormesh(suBZ/4.,cmap="coolwarm")
ax.set_aspect(1.0)
ticks = [0,suBZ.shape[0]/2,suBZ.shape[0]]
tick_labels = [r"$0$",r"$\pi/2$",r"$\pi$"]
ax.set_xticks(ticks)
ax.set_xticklabels(tick_labels,fontsize =18 )
ax.set_yticks(ticks)
ax.set_yticklabels(tick_labels,fontsize=18)
plt.title(r"$\chi^s(\mathbf{q})$, $n=0.82$",fontsize=22) 
plt.colorbar(cf)
plt.savefig("../plots/susc_q_0.82.png",dpi =300)

```

```python
def plot_susc(filename,dens,outfile="../plots/susc_q_"):
    plt.clf()
    suBZ = np.loadtxt(filename)
    ax = plt.subplot()
    cf = ax.pcolormesh(suBZ/4.,cmap="coolwarm")
    ax.set_aspect(1.0)
    ticks = [0,suBZ.shape[0]/2,suBZ.shape[0]]
    tick_labels = [r"$0$",r"$\pi/2$",r"$\pi$"]
    ax.set_xticks(ticks)
    ax.set_xticklabels(tick_labels,fontsize =18 )
    ax.set_yticks(ticks)
    ax.set_yticklabels(tick_labels,fontsize=18)
    plt.title(r"$\chi^s(\mathbf{q})$, $n=$"+dens,fontsize=22) 
    plt.colorbar(cf)
    plt.savefig(outfile+dens+".png",dpi =300)
def plot_bubble(filename,dens):
    plt.clf()
    suBZ = np.loadtxt(filename)
    ax = plt.subplot()
    cf =ax.pcolormesh(suBZ/4.,cmap="coolwarm")
    ax.set_aspect(1.0)
    ticks = [0,suBZ.shape[0]/2,suBZ.shape[0]]
    tick_labels = [r"$0$",r"$\pi/2$",r"$\pi$"]
    ax.set_xticks(ticks)
    ax.set_xticklabels(tick_labels,fontsize =18 )
    ax.set_yticks(ticks)
    ax.set_yticklabels(tick_labels,fontsize=18)
    plt.title(r"$\chi_0^s(\mathbf{q})$, $n=$"+dens,fontsize=22) 
    plt.colorbar(cf)
    plt.savefig("../plots/bubb_q_"+dens+".png",dpi =300)
```

```python
filename = "../data/incommensurate_susc/n0.82/suscBZ.txt"
filename_bubble = "../data/incommensurate_susc/n0.82/bubbleBZ.txt"
plot_susc(filename,"0.82")
```

```python
plot_bubble(filename_bubble,"0.82")
```

```python
filename = "../data/incommensurate_susc/n0.72/suscBZ.txt"
filename_bubble = "../data/incommensurate_susc/n0.72/bubbleBZ.txt"
plot_susc(filename,"0.72")
```
