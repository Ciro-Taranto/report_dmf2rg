```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mag_data = np.loadtxt('../data/HFU16t_max_mag.dat')
print mag_data.shape
# Multiply by 4 the mag channel
mag_data[:,1]  = 4.*mag_data[:,1]
```

```python
chi_data = np.loadtxt('../data/HFU16t_max_chi_mag.dat')
print mag_data.shape
# Devide by 4 the mag susc
chi_data[:,1]  = chi_data[:,1]/4.
```

```python
plt.clf()

plt.semilogx( mag_data[:,0] , mag_data[:,1], 'o-' )
plt.title(r'$\mathcal{M}$    $\mathbf{Q}=(\pi,\pi)$    $\Omega=0$',fontsize=18)
plt.xlabel(r'$\Lambda$',fontsize=18)

plt.gca().grid(linestyle='--')
```

```python
plt.clf()

plt.semilogx( chi_data[:,0] , chi_data[:,1], 'o-' )
plt.title(r'$\chi_{s}$    $\mathbf{Q}=(\pi,\pi)$    $\Omega=0$',fontsize=18)
plt.xlabel(r'$\Lambda$',fontsize=18)

plt.gca().grid(linestyle='--')
```
