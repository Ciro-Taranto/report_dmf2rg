```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

## $d$-wave at low temperature

```python
# In order: 0lam 1Uhub 2beta 3tpri 4tsec 5fill 6mag 7den_max 8den_min 9swave 10dwave
dwave_flow = np.loadtxt("../data/dwave_LT_rawdata_n0.820_tpri_-0.050_U2.00_beta90.00.dat")

# Unit: rescale by 4.
dwave_flow [:,2] =      dwave_flow[:,2] / 4.
for i in [1,3,4,6,7,8,9,10]:
    dwave_flow[:,i] = 4.*dwave_flow[:,i]
    
# get parameters
uhub, beta, tpri, tsec, dens = (list(set(dwave_flow[:,1]))[0], list(set(dwave_flow[:,2]))[0],  
                                list(set(dwave_flow[:,3]))[0], list(set(dwave_flow[:,4]))[0], 
                                list(set(dwave_flow[:,5]))[0])
```

```python
plt.clf()

ax = plt.subplot()

fs = 18 
title = r"$T="+ "{:.3f}".format(1./beta)+"t$, $U="+"%1.0f"%(uhub)+"t$"
label = r"$t'="+"{:.2f}".format(tpri)+"$t, $n="+"{:.3f}".format(dens)+"$"
ax.plot((dwave_flow[:,0]),dwave_flow[:,10],marker='o', color="blue",label=label,alpha =0.5)
ax.legend(fontsize = fs)
ax.set_title(title,fontsize =fs +2)
ax.set_xscale('log')
ax.set_xlabel(r"$\Lambda$",fontsize = fs+2)
ax.set_ylabel(r"$\mathrm{max}(\mathcal{D}^\Lambda)$",fontsize = fs+2)
ax.set_xticks([1.0, 0.5,0.2,0.1,0.05])
ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax.grid()
```

## $d$-wave from perpendicular ladder

```python
# 0lam 1beta 2fill/2 3tpri 4max_mag 5mag_pp 6dwave
dwave_flow = np.loadtxt("../data/dwave_PL_n0.82-tpri-minus0.05-U2.0-b90.0.dat")

uhub,beta,tpri,dens = 8.,90/4.,-0.05*4.,0.82

# Unit: rescale by 4.
dwave_flow [:,1] =      dwave_flow[:,1] / 4.
for i in [3,4,6]:
    dwave_flow[:,i] = 4.*dwave_flow[:,i]
```

```python
plt.clf()

ax = plt.subplot()

fs = 18 
title = r"$T="+ "{:.3f}".format(1./beta)+"t$, $U="+"%1.0f"%(uhub)+"t$"
label = r"$t'="+"{:.2f}".format(tpri)+"$t, $n="+"{:.3f}".format(dens)+"$"
ax.semilogx((dwave_flow[:,0]),dwave_flow[:,6],marker='o', color="blue",label=r'$\mathcal{D}$',alpha = 0.5)
ax.semilogx((dwave_flow[:,0]),dwave_flow[:,4],marker='o', color="red",label=r'$\mathcal{M}$',alpha = 0.5)
plt.gca().grid(linestyle='--')
ax.legend(fontsize = fs)
ax.set_title(title,fontsize =fs +2)
#ax.set_xscale('log')
ax.set_xlabel(r"$\Lambda$",fontsize = fs+2)
#ax.set_ylabel(r"$\mathrm{max}(\mathcal{D}^\Lambda)$",fontsize = fs+2)
#ax.set_xticks([1.0, 0.5,0.2,0.1,0.05])
#ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
```
