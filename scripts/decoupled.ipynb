```python
import h5py 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
rawdata = np.loadtxt('../data/rawdata_decoupled_phasediag.dat')
fillings= np.unique(rawdata[:,4])
criData = []
for fill in fillings:
        print (fill)
        atFill = rawdata[np.where(rawdata[:,4]==fill)]
        lmin = np.argmin(atFill[:,0])
        print (lmin)
        criData.append(atFill[lmin])
criData = np.array(criData)
print (criData.shape)
print (criData[:,0])
print (criData[:,4])
print(criData[:,11])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax2=plt.twinx()
rawdata = np.loadtxt("../data/rawdata_phasediag.dat")
ln1=ax.plot(1-rawdata[:,4],rawdata[:,0],'-',label=r"$\Lambda_{cri}$")
ax.fill_between(1-rawdata[:,4],rawdata[:,0],alpha=0.2)
ax.scatter(1-rawdata[:,4],rawdata[:,0],color="steelblue",alpha=0.7,edgecolor="black",s=70)
ax.scatter(1-criData[:,4],criData[:,0],color="orange",alpha=0.7,edgecolor="black",s=70)
ax.plot(1-criData[:,4],criData[:,0],'-',label=r"$\Lambda_{cri}$ decoup")
ax.fill_between(1-criData[:,4],criData[:,0],alpha=0.2)
ax.set_ylim([0,0.45])
ln2=ax2.plot(1-rawdata[:,4],4*rawdata[:,11],"-",color="green",linestyle="dashed",label=r"$\mathcal{D}^{\mathbf{Q}=\mathbf{0},\Omega=0}$")
ax2.scatter(1-rawdata[:,4],4*rawdata[:,11],marker='*',color="green",linestyle="dashed",s=150)
ax2.plot(1-criData[:,4],4*criData[:,11],"-",color="purple",linestyle="dashed",label=r"$\mathcal{D}^{\mathbf{Q}=\mathbf{0},\Omega=0}$")
ax2.scatter(1-criData[:,4],4*criData[:,11],marker='d',color="purple",linestyle="dashed",s=150)




ax2.set_ylim([0,70])
ax2.set_ylabel(r"$\mathcal{D}^{\mathbf{Q}=\mathbf{0},\Omega=0}$",fontsize=30)
ax.set_ylabel(r"$\Lambda_{cri}$",fontsize=30)
ax2.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.annotate(r"$AF$",xy=(0.06,0.150),fontsize=40)
ax.set_xlabel(r"$\delta$",fontsize=30)

lns = ln1+ln2
labs = [l.get_label() for l in lns]
#ax.legend(lns, labs, loc=0,fontsize=20)
#plt.ylabel(r"")
ax.set_xlim([0.02,0.28])
plt.tight_layout()
plt.savefig("../plots/pairing_plus_criscale.png",dpi=300)
```
